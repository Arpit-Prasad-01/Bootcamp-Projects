```python
# Data Analysis with Pandas & Visualization via Matplotlib
# tudent Records & Gapminder Analysis

# Visualise dataset using matplotlib library

import pandas as pd

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [12000, 15000, 18000, 17000, 22000, 25000,
              28000, 27000, 26000, 30000, 35000, 40000],
    'Profit': [2500, 3000, 3500, 3200, 4500, 5000,
               6000, 5800, 5500, 6500, 8000, 9500],
    'Customers': [150, 180, 220, 210, 260, 290,
                  330, 320, 310, 350, 420, 500],
    'Returns': [12, 15, 18, 14, 20, 25,
                22, 21, 19, 24, 30, 35]
}

df = pd.DataFrame(data)

print(df.head())
```

```python
# Line Chart
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Sales'], marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(8,5))

plt.bar(df['Month'], df['Profit'])

plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit ($)")
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8,5))

plt.scatter(df['Customers'], df['Sales'])

plt.title("Customers vs Sales")
plt.xlabel("Number of Customers")
plt.ylabel("Sales")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8,5))

plt.hist(df['Sales'], bins=6)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pie chart
quarters = {
    'Q1': sum(df['Sales'][0:3]),
    'Q2': sum(df['Sales'][3:6]),
    'Q3': sum(df['Sales'][6:9]),
    'Q4': sum(df['Sales'][9:12])
}

plt.figure(figsize=(7,7))

plt.pie(quarters.values(),
        labels=quarters.keys(),
        autopct='%1.1f%%')

plt.title("Sales Contribution by Quarter")
plt.show()
```

```python
# Box plot
plt.figure(figsize=(8,5))

plt.boxplot([df['Sales'],
             df['Profit'],
             df['Returns']],
            labels=['Sales','Profit','Returns'])

plt.title("Distribution of Business Metrics")
plt.show()
```

```python
# Horizontal Bar Chart
top = df.sort_values('Sales', ascending=False)

plt.figure(figsize=(8,5))

plt.barh(top['Month'], top['Sales'])

plt.title("Sales by Month")
plt.xlabel("Sales")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)


# Check column names
print(df.columns)

# Calculate average life expectancy by continent
avg_life_exp = df.groupby('continent')['lifeExp'].mean()

# Create bar chart
plt.figure(figsize=(8, 5))
avg_life_exp.plot(kind='bar')

plt.title('Average Life Expectancy by Continent')
plt.xlabel('Continent')
plt.ylabel('Average Life Expectancy')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df = pd.read_csv(url)

data = df.groupby("continent")["lifeExp"].mean()

plt.bar(data.index, data.values)

plt.figure(figsize=(8,5))
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")
plt.title("Average Life Expectancy vs Continent")

plt.show()
```

```python

data = df[df["country"].isin(["India","China","United States"])]

plt.figure(figsize=(8,5))

for country in data["country"].unique():
    temp = data[data["country"] == country]
    plt.plot(temp["year"], temp["lifeExp"], marker="o", label=country)

plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.title("Life Expectancy Over Time")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["gdpPercap"], df["lifeExp"])

plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.title("GDP per Capita vs Life Expectancy")

plt.show()
```

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/session4_dummy_students_data.csv (1).xls")

print(df.head())


df['average'] = (df['math'] + df['science'] + df['english']) / 3

# Pass aur Fail check by looping
df['result'] = ['Pass' if x >= 40 else 'Fail' for x in df['average']]
print(df[['name', 'average', 'result']].head())

#total average marks
class_avg = df['average'].mean()
print("\nClass Average is:", class_avg)


print("Math Max:", df['math'].max(), "Min:", df['math'].min())
print("Science Max:", df['science'].max(), "Min:", df['science'].min())
print("English Max:", df['english'].max(), "Min:", df['english'].min())


max_marks = df['average'].max()
top_student = df[df['average'] == max_marks]
print(top_student[['name', 'average']])


low_attendance = df[df['attendance_percent'] < 75]
print(low_attendance[['name', 'attendance_percent']])


less_assignments = df[df['assignments_submitted'] < 6]
print(less_assignments[['name', 'assignments_submitted']])


url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
data = pd.read_csv(url)

print("Data Shape:", data.shape)
print("Data Columns Name:", data.columns)

print("Total Countries:", data['country'].nunique())
print("Total Continents:", data['continent'].nunique())


print("\nContinent wise count:")
print(data['continent'].value_counts())


print("\nNull values check:")
print(data.isnull().sum())


print("\nAverage Life Exp:", data['lifeExp'].mean())
print("Average GDP:", data['gdpPercap'].mean())


highest_gdp = data.loc[data['gdpPercap'].idxmax()]
print(highest_gdp)


# Plot 1: Bar chart for continent wise average life expectancy
bar_data = data.groupby("continent")["lifeExp"].mean()
plt.bar(bar_data.index, bar_data.values, color='teal') # Simple teal color use kiya hai
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")
plt.title("Average Life Expectancy vs Continent")
plt.show()

# Plot 2: Line chart comparing India, China and United States
countries = data[data["country"].isin(["India", "China", "United States"])]
plt.figure()
for c in countries["country"].unique():
    temp_df = countries[countries["country"] == c]
    plt.plot(temp_df["year"], temp_df["lifeExp"], marker="o", label=c)

plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.title("Life Expectancy Over Time")
plt.legend()
plt.show()

# Plot 3: Scatter plot for GDP vs Life Expectancy
plt.figure()
plt.scatter(data["gdpPercap"], data["lifeExp"], color='red', alpha=0.5) # simple red scatter points
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.title("GDP per Capita vs Life Expectancy")
plt.show()
```
